```python
import torch
import numpy as np

resume = 'binput-prerprelu-pg-resnet20.pt'
loaded_dict = torch.load(resume)
```

```python
from collections import OrderedDict
new_dict = OrderedDict()
for k, v in loaded_dict.items():
    if 'tracked' not in k:
        new_dict[k] = v
new_dict['encoder.placeholder'] = loaded_dict['encoder.placeholder'][0:1, :, :, :, :]
```

```python

for k, v in new_dict.items():
    v = v.cpu().numpy()
    if 'relu' in k or 'threshold' in k:
        v = v.flatten()
    new_dict[k] = v
    print(k)
    print(v.shape)
```

```python
def fuse_bn(weight, bias, mean, var):
    fused_weight = weight/np.sqrt(var - 1e-5)
    fused_bias = bias - mean*weight/np.sqrt(var - 1e-5)
    return fused_weight, fused_bias
```
